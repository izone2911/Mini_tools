```python
# demo move move subdirectories to parent and delete empty needed folder
import os
import shutil

def move_subdirectories_to_parent(parent_directory):
    # Lp qua tt c cc mc trong th mc cha
    for item in os.listdir(parent_directory):
        item_path = os.path.join(parent_directory, item)
        print(item_path)
        # Kim tra xem mc c phi l th mc khng
        if os.path.isdir(item_path):
            # Lp qua tt c cc th mc con trong th mc cha
            for sub_item in os.listdir(item_path):
                file_extension = os.path.splitext(sub_item)[1]  
                sub_item_path = os.path.join(item_path, sub_item)
                if file_extension != '.bmp':
                    continue
                print(file_extension,end='---')
                print(sub_item,end='')
                shutil.move(sub_item_path, parent_directory)
                print(f" di chuyn: {sub_item_path} ra ngoi {parent_directory}")

# Thay th 'path_to_parent_directory' bng ng dn n th mc cha
parent_directory = './CASIA1 - Copy/CASIA Iris Image Database (version 1.0)'
for i in range(1,109):
    j = str(i)
    while len(j) < 3:
        j = '0' + j
    new_parent_directory = os.path.join(parent_directory, j)
    move_subdirectories_to_parent(new_parent_directory)
```
